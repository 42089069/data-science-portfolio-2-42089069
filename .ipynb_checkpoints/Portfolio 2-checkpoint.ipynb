This portfolio reproduces some of the analysis on data from the paper

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.feature_selection import RFE, RFECV
from sklearn.metrics import accuracy_score
%matplotlib inline
```

## Parsing the Data
The data for this analysis was provided by the authors of the original paper, along with other data sets and scripts used in their analysis, from their github repository https://github.com/LuisM78/Appliances-energy-prediction-data

```python
df = pd.read_csv('data/energydata_complete.csv', 
                 index_col = 0,
                parse_dates=['date'])
df.head()
```

```python
training = pd.read_csv('data/training.csv', 
                        index_col = 0,
                        parse_dates=['date'])
training.head()
```

```python
testing = pd.read_csv('data/testing.csv', 
                        index_col = 0,
                        parse_dates=['date'])
testing.head()
```

## Exploring the data

### Reproducing Figure 7
Figure 7 shows the energy consumed by appliances over the period of study. Energy is measured in watt hours (Wh). The total time span is 137 days, with each recorded entry being made at ten minute intervals.

```python
fig, df_ax = plt.subplots(figsize=(30,10))
df_ax.xaxis.set_major_locator(mdates.MonthLocator())
df_ax.set_xlabel('', fontsize = 20)
df_ax.set_ylabel('Appliances(Wh)', fontsize = 20)
df.index.name = 'Time'
df.plot(y='Appliances',
        kind = 'line',
        yticks = np.arange(0, 1000, 300),
        fontsize = 20,
        ax = df_ax)

fig, df2_ax = plt.subplots(figsize=(30,10))
df2_ax.xaxis.set_major_locator(mdates.DayLocator())
df2_ax.set_xlabel('', fontsize = 20)
df2_ax.set_ylabel('Appliances(Wh)', fontsize = 20)
df.index.name = 'Time (Week 1)'
df.iloc[0:1008].plot(y='Appliances',
                     kind = 'line',
                     yticks = np.arange(0, 1000, 300),
                     fontsize = 20,
                     ax = df2_ax)

df.index.name = 'date'
```

### Reproducing figure 9

Figure 9 is intended to represent the dependency between the variable Appliances and other variables from the data set. In the lower triangle bivariate scatter plot represent the strength of the relationship between the two variables. The red line represents the linear regression fits. The diagonal plots the variable's frequency scores through histogram plots. In the upper triangle contains the Pearson correlation scores of two variables. 

```python
def corr(x, y, **kwargs):
    coef = np.corrcoef(x, y)[0][1]
    label = str(round(coef, 2))
    ax = plt.gca()
    ax.annotate(label, xy = (0.45, 0.5), size = 20, xycoords = ax.transAxes)

trainplot = sns.PairGrid(training, vars=['Appliances', 'lights', 'T1', 'RH_1', 'T2', 'RH_2', 'T3', 'RH_3'])
trainplot.map_diag(plt.hist)
trainplot.map_lower(sns.regplot, line_kws={'color': 'red'})
trainplot.map_upper(corr)
```

### Reproducing figure 10

Figure 10 is a breakdown of the hourly consumption of power by appliances over a given week. Within the study the week represented was not specified, so in this reproduction the first week recorded was selected.

```python
heatdf = pd.DataFrame(index=list(range(24)), columns=list(range(7)))
heatdf = heatdf.fillna(0)

for index, row in df['Appliances'].iloc[0:1008].iteritems():
    heatdf.at[index.hour, index.weekday()] = heatdf.at[index.hour, index.weekday()] + row

figHeat, axHeat = plt.subplots(figsize=(15,15))
sns.heatmap(heatdf, square =True, ax=axHeat)
```

## Fitting a linear model

```python
x = training.drop('Appliances', axis=1)
y = training[['Appliances']]
testx = testing.drop('Appliances', axis=1)
testy = testing[['Appliances']]

x['Day_of_week'] = x['Day_of_week'].astype('category')
x['Day_of_week'] = x['Day_of_week'].cat.reorder_categories(x['Day_of_week'].unique(), ordered=True)
x['Day_of_week'] = x['Day_of_week'].cat.codes

x['WeekStatus'] = x['WeekStatus'].astype('category')
x['WeekStatus'] = x['WeekStatus'].cat.reorder_categories(x['WeekStatus'].unique(), ordered=True)
x['WeekStatus'] = x['WeekStatus'].cat.codes

testx['Day_of_week'] = testx['Day_of_week'].astype('category')
testx['Day_of_week'] = testx['Day_of_week'].cat.reorder_categories(testx['Day_of_week'].unique(), ordered=True)
testx['Day_of_week'] = testx['Day_of_week'].cat.codes

testx['WeekStatus'] = testx['WeekStatus'].astype('category')
testx['WeekStatus'] = testx['WeekStatus'].cat.reorder_categories(testx['WeekStatus'].unique(), ordered=True)
testx['WeekStatus'] = testx['WeekStatus'].cat.codes
```

```python
regModel = LinearRegression()
regModel.fit(x,y)
pred = regModel.predict(testx)
MSE = mean_squared_error(testy, pred)
print("MSE:", MSE)
print("RMSE", np.sqrt(MSE))
print("R^2:", r2_score(testy, pred))
```

## Applying Recursive Feature Estimation

```python
selector = RFE(regModel, 31, step=1)
selector = selector.fit(x, y.values.ravel())
Y_pred = selector.predict(testx)

MSE = mean_squared_error(testy, Y_pred)
print("MSE:", MSE)
print("RMSE", np.sqrt(MSE))
print("R^2:", r2_score(testy, Y_pred))
```
